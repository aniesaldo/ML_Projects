3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

```python
#3.C)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

p = 3

x_values = np.arange(0, 16)

cum_dist_func = poisson.cdf(x_values, mu=p)

print('Answer 3C:'+'\n')
plt.figure(figsize=(5, 3))
plt.plot(x_values, cum_dist_func, marker='o', linestyle='-')
plt.title('Cumulative Probability of Cars Sold per Week')
plt.xlabel('Number of Cars Sold')
plt.ylabel('Cumulative Probability')
plt.xticks(x_values)
plt.show()
```

```python
def histogram_boxplot(data, feature, figsize=(12, 7), kde=True, bins=20):
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,
        sharex=True,
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize)
    sns.boxplot(data=data, x=feature, ax=ax_box2, showmeans=True)
    
    if bins:
      sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins)
    else: 
      sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2)
    
    ax_hist2.axvline(data[feature].mean(), color="green", linestyle="--")  # Mean for histogram
    
    ax_hist2.axvline(data[feature].median(), color="black", linestyle="-")  # Median for histogram
```

B. Write the null hypothesis and alternative hypothesis.


Null Hypothesis (H0):
    
There is no significant difference in the distribution of funds raised among companies in different operating states.

Alternative Hypothesis (H1):

There is a significant difference in the distribution of funds raised among companies in different operating states

C. Test for significance and conclusion:
    

```python
from scipy.stats import f_oneway

# for Op_state, grouped in data3.groupby('OperatingState'):
#     print(Op_state)
#     print(grouped['Funds_in_million'])

funds_by_Opstate=[grouped['Funds_in_million'] for Op_state, grouped in data3.groupby('OperatingState')]

fstatistic, p_value = f_oneway(*funds_by_Opstate)
print("F-statistic:", fstatistic)
print("p-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in the distribution of funds raised among different operating states.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the distribution of funds raised among different operating states.")
```
